```python
#OPERACIONES BASICAS DE PROGRAMACION
N1= 6
N2= 10
Suma=N1 + N2
Resta=N1 + N2
Division=N1 / N2
Multiplicacion=N1 * N2
Operacion= (N1 + N2) + (N1 / N2)
Potencia= N1 **  N2 #el doble asterico es para potencia

print("El valor de la operacion es igual a:", Operacion)
print("El valor de la suma es igual a:", Suma)
print("El valor de la resta es igual a:", Resta)
print("El valor de la division es igual a:", Division)
print("El valor de la multiplicacion es igual a:", Multiplicacion)
print("El valor de la potencia es igual a:", Potencia)
```
